```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from rdkit import Chem, RDLogger
from rdkit.Chem import Descriptors, Draw, AllChem
from rdkit.Chem.Scaffolds import MurckoScaffold
from rdkit.Chem import rdMolDescriptors

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

RDLogger.DisableLog('rdApp.*')

COLORS = {'High (>7)': '#e74c3c', 'Medium (5-7)': '#f39c12', 'Low (<5)': '#3498db'}
print("All packages imported successfully.")
```

```
from google.colab import files
uploaded = files.upload()  # upload EGFR_full_features.csv when prompted

df = pd.read_csv('EGFR_full_features.csv')
df = df.dropna(subset=['pIC50'])
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['pIC50'])

print(f"Dataset loaded: {len(df)} compounds")
print(f"pIC50 range: {df['pIC50'].min():.2f} – {df['pIC50'].max():.2f}")
print(f"pIC50 mean ± std: {df['pIC50'].mean():.2f} ± {df['pIC50'].std():.2f}")
```

```
df['activity_class'] = pd.cut(
    df['pIC50'],
    bins=[0, 5, 7, 20],
    labels=['Low (<5)', 'Medium (5-7)', 'High (>7)']
)

class_counts = df['activity_class'].value_counts()
print("Activity class distribution:")
print(class_counts)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax = axes[0]
for cls, color in COLORS.items():
    subset = df[df['activity_class'] == cls]['pIC50']
    ax.hist(subset, bins=40, color=color, alpha=0.7, label=cls, edgecolor='white', linewidth=0.3)
ax.axvline(5, color='gray', linestyle='--', lw=1.5, label='Threshold @ pIC50=5')
ax.axvline(7, color='black', linestyle='--', lw=1.5, label='Threshold @ pIC50=7')
ax.set_xlabel('pIC50', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('pIC50 Distribution by Activity Class', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

ax2 = axes[1]
wedge_colors = [COLORS[c] for c in class_counts.index]
ax2.pie(class_counts.values, labels=class_counts.index, colors=wedge_colors,
        autopct='%1.1f%%', startangle=90, textprops={'fontsize': 11})
ax2.set_title('Activity Class Proportions', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('step4_activity_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_activity_distribution.png")
```

```
key_descriptors = {
    'MolWt':             'Molecular Weight (Da)',
    'MolLogP':           'LogP (Lipophilicity)',
    'NumHDonors':        'H-Bond Donors',
    'NumHAcceptors':     'H-Bond Acceptors',
    'TPSA':              'TPSA (Å²)',
    'NumRotatableBonds': 'Rotatable Bonds',
    'RingCount':         'Ring Count',
    'qed':               'QED Drug-likeness'
}

df_hl = df[df['activity_class'].isin(['High (>7)', 'Low (<5)'])].copy()

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

for idx, (col, label) in enumerate(key_descriptors.items()):
    ax = axes[idx]
    if col not in df_hl.columns:
        ax.set_visible(False)
        continue
    for cls in ['High (>7)', 'Low (<5)']:
        subset = df_hl[df_hl['activity_class'] == cls][col].dropna()
        q1, q99 = subset.quantile(0.01), subset.quantile(0.99)
        subset = subset.clip(q1, q99)
        ax.hist(subset, bins=30, color=COLORS[cls], alpha=0.65,
                label=cls, edgecolor='white', linewidth=0.3)
    ax.set_xlabel(label, fontsize=10)
    ax.set_ylabel('Count', fontsize=10)
    ax.set_title(label, fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.suptitle('Physicochemical Descriptor Comparison: High vs Low Activity EGFR Inhibitors',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('step4_descriptor_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_descriptor_comparison.png")
```

```
summary_rows = []
for col, label in key_descriptors.items():
    if col not in df.columns:
        continue
    for cls in ['High (>7)', 'Low (<5)', 'Medium (5-7)']:
        vals = df[df['activity_class'] == cls][col].dropna()
        summary_rows.append({
            'Descriptor': label,
            'Activity Class': cls,
            'Mean': vals.mean(),
            'Median': vals.median(),
            'Std': vals.std()
        })

summary_df = pd.DataFrame(summary_rows)
pivot = summary_df.pivot_table(
    index='Descriptor', columns='Activity Class', values='Mean'
).round(2)
print("\n=== Mean Descriptor Values by Activity Class ===")
print(pivot.to_string())
pivot.to_csv('step4_descriptor_summary.csv')

# Box plots
top_desc   = ['MolWt', 'MolLogP', 'TPSA', 'qed']
top_labels = ['Molecular Weight (Da)', 'LogP', 'TPSA (Å²)', 'QED Drug-likeness']
order   = ['Low (<5)', 'Medium (5-7)', 'High (>7)']
palette = {k: v for k, v in COLORS.items()}

fig, axes = plt.subplots(1, 4, figsize=(18, 6))
for idx, (col, label) in enumerate(zip(top_desc, top_labels)):
    if col not in df.columns:
        continue
    plot_data = df[df['activity_class'].isin(order)][[col, 'activity_class']].dropna()
    q1, q99 = plot_data[col].quantile(0.01), plot_data[col].quantile(0.99)
    plot_data[col] = plot_data[col].clip(q1, q99)
    ax = axes[idx]
    sns.boxplot(data=plot_data, x='activity_class', y=col,
                order=order, palette=palette, ax=ax,
                flierprops=dict(marker='o', markersize=2, alpha=0.3))
    ax.set_xlabel('Activity Class', fontsize=11)
    ax.set_ylabel(label, fontsize=11)
    ax.set_title(label, fontsize=12, fontweight='bold')
    ax.tick_params(axis='x', rotation=15)
    ax.grid(True, axis='y', alpha=0.3)

plt.suptitle('SAR: Key Descriptors Across Activity Classes', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('step4_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_boxplots.png")
```

```
def get_scaffold(smiles):
    mol = Chem.MolFromSmiles(str(smiles))
    if mol is None:
        return None
    try:
        scaffold = MurckoScaffold.GetScaffoldForMol(mol)
        return Chem.MolToSmiles(scaffold, isomericSmiles=False) if scaffold else None
    except:
        return None

print("Extracting Murcko scaffolds... (~60 seconds)")
df['scaffold'] = df['clean_smiles'].apply(get_scaffold)
df_sc = df.dropna(subset=['scaffold']).copy()

print(f"Compounds with valid scaffolds: {len(df_sc)}")
print(f"Unique scaffolds: {df_sc['scaffold'].nunique()}")

scaffold_stats = df_sc.groupby('scaffold').agg(
    count=('pIC50', 'count'),
    mean_pIC50=('pIC50', 'mean'),
    max_pIC50=('pIC50', 'max'),
    std_pIC50=('pIC50', 'std')
).reset_index().sort_values('count', ascending=False)

print("\nTop 10 most frequent scaffolds:")
print(scaffold_stats.head(10)[['scaffold','count','mean_pIC50','max_pIC50']].to_string())

top_active = scaffold_stats[scaffold_stats['count'] >= 5].sort_values('mean_pIC50', ascending=False)
print("\nTop 10 most ACTIVE scaffold families (n≥5):")
print(top_active.head(10)[['scaffold','count','mean_pIC50','max_pIC50']].to_string())
```

```
substructures = {
    'Quinazoline core':        'c1cnc2ccccc2n1',
    'Anilinoquinazoline':      'c1ccc(Nc2ncnc3ccccc23)cc1',
    'Acrylamide warhead':      'C=CC(=O)N',
    'Aminopyrimidine (hinge)': 'c1cnc(N)nc1',
    'Halogen (F/Cl/Br)':       '[F,Cl,Br]',
    'Pyridine ring':           'c1ccncc1',
    'Morpholine':              'C1COCCN1',
    'Sulfonamide':             'S(=O)(=O)N'
}

def has_substructure(smiles, smarts):
    mol = Chem.MolFromSmiles(str(smiles))
    pattern = Chem.MolFromSmarts(smarts)
    if mol is None or pattern is None:
        return False
    return mol.HasSubstructMatch(pattern)

print("Computing substructure presence... (~30s)")
for name, smarts in substructures.items():
    col = f'has_{name.replace(" ","_").replace("/","_").replace("(","").replace(")","")}'
    df[col] = df['clean_smiles'].apply(lambda s: has_substructure(s, smarts))

results = []
for name, smarts in substructures.items():
    col = f'has_{name.replace(" ","_").replace("/","_").replace("(","").replace(")","")}'
    with_ss    = df[df[col] == True]['pIC50']
    without_ss = df[df[col] == False]['pIC50']
    results.append({
        'Substructure':          name,
        'N (with)':              len(with_ss),
        'N (without)':           len(without_ss),
        'Mean pIC50 (with)':     with_ss.mean(),
        'Mean pIC50 (without)':  without_ss.mean(),
        'Delta pIC50':           with_ss.mean() - without_ss.mean()
    })

sar_ss_df = pd.DataFrame(results).sort_values('Delta pIC50', ascending=False)
print("\n=== Substructure SAR Table ===")
print(sar_ss_df.round(3).to_string(index=False))
```

```
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

ax = axes[0]
colors_bar = ['#e74c3c' if d > 0 else '#3498db' for d in sar_ss_df['Delta pIC50']]
bars = ax.barh(sar_ss_df['Substructure'], sar_ss_df['Delta pIC50'],
               color=colors_bar, edgecolor='gray', linewidth=0.5)
ax.axvline(0, color='black', lw=1)
ax.set_xlabel('ΔpIC50 (with − without substructure)', fontsize=11)
ax.set_title('Substructure Impact on EGFR Inhibitory Potency', fontsize=12, fontweight='bold')
ax.grid(True, axis='x', alpha=0.3)
for bar, val in zip(bars, sar_ss_df['Delta pIC50']):
    ax.text(bar.get_width() + (0.02 if val >= 0 else -0.02),
            bar.get_y() + bar.get_height()/2,
            f'{val:+.2f}', va='center',
            ha='left' if val >= 0 else 'right', fontsize=9)

ax2 = axes[1]
x = np.arange(len(sar_ss_df))
w = 0.35
ax2.bar(x - w/2, sar_ss_df['Mean pIC50 (with)'],    w, label='With substructure',    color='#e74c3c', alpha=0.8, edgecolor='gray')
ax2.bar(x + w/2, sar_ss_df['Mean pIC50 (without)'], w, label='Without substructure', color='#3498db', alpha=0.8, edgecolor='gray')
ax2.set_xticks(x)
ax2.set_xticklabels(sar_ss_df['Substructure'], rotation=35, ha='right', fontsize=9)
ax2.set_ylabel('Mean pIC50', fontsize=11)
ax2.set_title('Mean pIC50: With vs Without Key Substructures', fontsize=12, fontweight='bold')
ax2.axhline(7, color='green', linestyle='--', lw=1.2, label='High-activity threshold')
ax2.legend(fontsize=9)
ax2.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('step4_substructure_sar.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_substructure_sar.png")
```

```
pca_descriptors = ['MolWt','MolLogP','NumHDonors','NumHAcceptors',
                   'TPSA','NumRotatableBonds','RingCount','qed',
                   'MaxAbsEStateIndex','MinEStateIndex',
                   'FpDensityMorgan1','FpDensityMorgan2',
                   'BalabanJ','BertzCT','HallKierAlpha']
pca_descriptors = [c for c in pca_descriptors if c in df.columns]

df_pca = df[pca_descriptors + ['pIC50','activity_class']].dropna().copy()
for col in pca_descriptors:
    q1, q99 = df_pca[col].quantile(0.01), df_pca[col].quantile(0.99)
    df_pca[col] = df_pca[col].clip(q1, q99)

X_scaled = StandardScaler().fit_transform(df_pca[pca_descriptors].values)
pca = PCA(n_components=min(len(pca_descriptors), 10))
X_pca_coords = pca.fit_transform(X_scaled)
ev = pca.explained_variance_ratio_
y_pca = df_pca['pIC50'].values

print(f"PC1: {ev[0]*100:.1f}%  |  PC2: {ev[1]*100:.1f}%  |  Combined: {(ev[0]+ev[1])*100:.1f}%")

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

ax = axes[0]
scatter = ax.scatter(X_pca_coords[:,0], X_pca_coords[:,1],
                     c=y_pca, cmap='RdYlGn', alpha=0.65, s=12,
                     edgecolors='none', vmin=4, vmax=10)
plt.colorbar(scatter, ax=ax, shrink=0.85).set_label('pIC50', fontsize=11)
ax.set_xlabel(f'PC1 ({ev[0]*100:.1f}% variance)', fontsize=12)
ax.set_ylabel(f'PC2 ({ev[1]*100:.1f}% variance)', fontsize=12)
ax.set_title('PCA Chemical Space — Colored by pIC50', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.25)

ax2 = axes[1]
for cls in ['Low (<5)', 'Medium (5-7)', 'High (>7)']:
    mask = df_pca['activity_class'] == cls
    ax2.scatter(X_pca_coords[mask,0], X_pca_coords[mask,1],
                c=COLORS[cls], label=cls, alpha=0.6, s=12, edgecolors='none')
ax2.set_xlabel(f'PC1 ({ev[0]*100:.1f}% variance)', fontsize=12)
ax2.set_ylabel(f'PC2 ({ev[1]*100:.1f}% variance)', fontsize=12)
ax2.set_title('PCA Chemical Space — Colored by Activity Class', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10, markerscale=2)
ax2.grid(True, alpha=0.25)

plt.suptitle('EGFR Inhibitor Chemical Space (PCA)', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('step4_pca_chemical_space.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_pca_chemical_space.png")
```

```
MAX_TSNE = 3000

def get_morgan_fp(smiles, radius=2, nBits=1024):
    mol = Chem.MolFromSmiles(str(smiles))
    if mol is None:
        return np.zeros(nBits)
    return np.array(AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits))

df_tsne = df.dropna(subset=['pIC50','clean_smiles','activity_class']).copy()
if len(df_tsne) > MAX_TSNE:
    df_tsne = df_tsne.groupby('activity_class', group_keys=False).apply(
        lambda x: x.sample(min(len(x), int(MAX_TSNE * len(x) / len(df_tsne))), random_state=42)
    )
    print(f"Subsampled to {len(df_tsne)} compounds (stratified)")

print("Computing Morgan fingerprints...")
fps = np.array([get_morgan_fp(s) for s in df_tsne['clean_smiles']])

print("Running t-SNE... (~2 minutes)")
tsne = TSNE(n_components=2, perplexity=40, n_iter=1000,
            random_state=42, learning_rate='auto', init='pca')
tsne_coords = tsne.fit_transform(fps)
df_tsne['tsne_x'] = tsne_coords[:, 0]
df_tsne['tsne_y'] = tsne_coords[:, 1]
print("t-SNE complete!")
```

```
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

ax = axes[0]
sc = ax.scatter(df_tsne['tsne_x'], df_tsne['tsne_y'],
                c=df_tsne['pIC50'], cmap='RdYlGn',
                alpha=0.6, s=10, edgecolors='none', vmin=4, vmax=10)
plt.colorbar(sc, ax=ax, shrink=0.85).set_label('pIC50', fontsize=11)
ax.set_xlabel('t-SNE Dimension 1', fontsize=12)
ax.set_ylabel('t-SNE Dimension 2', fontsize=12)
ax.set_title('t-SNE Chemical Space — Colored by pIC50', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.2)

ax2 = axes[1]
for cls in ['Low (<5)', 'Medium (5-7)', 'High (>7)']:
    mask = df_tsne['activity_class'] == cls
    ax2.scatter(df_tsne.loc[mask,'tsne_x'], df_tsne.loc[mask,'tsne_y'],
                c=COLORS[cls], label=cls, alpha=0.55, s=10, edgecolors='none')
ax2.set_xlabel('t-SNE Dimension 1', fontsize=12)
ax2.set_ylabel('t-SNE Dimension 2', fontsize=12)
ax2.set_title('t-SNE Chemical Space — Colored by Activity Class', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10, markerscale=3)
ax2.grid(True, alpha=0.2)

plt.suptitle('EGFR Inhibitor Chemical Space (t-SNE on Morgan Fingerprints)',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('step4_tsne_chemical_space.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: step4_tsne_chemical_space.png")
```
